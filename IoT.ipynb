```python
# Using CSV File
import paho.mqtt.client as mqtt #import the client1
import time
import json
############
def on_message(client, userdata, message):
    print("message received " ,str(message.payload.decode("utf-8")))
    print("message topic=",message.topic)
    print("message qos=",message.qos)
    print("message retain flag=",message.retain)
########################################
broker_address="broker.mqttdashboard.com"
#broker_address="iot.eclipse.or
with open('myjson.json') as f:
    jsonObject = json.load(f)
    print(jsonObject)
print("creating new instance")
client = mqtt.Client("mytpha4") #create new instance
for key in jsonObject:
    client.on_message=on_message #attach function to callback
    print("connecting to broker")
    client.connect(broker_address) #connect to broker
    client.loop_start() #start the loop
    print("Subscribing to topic","temp/alert")
    client.subscribe("temp/alert")
    print("Publishing message to topic","temp/alert")
    client.publish("temp/alert",key["Temperature"])
    time.sleep(4) # wait
    client.loop_stop() #stop the loop
    client.on_message=on_message #attach function to callback
    print("connecting to broker")
    client.connect(broker_address) #connect to broker
    client.loop_start() #start the loop
    print("Subscribing to topic","pressure/alert")
    client.subscribe("pressure/alert")
    print("Publishing message to topic","pressure/alert")
    client.publish("pressure/alert",key["Pressure"])
    time.sleep(4) # wait
    client.loop_stop() #stop the loop
    client.on_message=on_message #attach function to callback
    print("connecting to broker")
    client.connect(broker_address) #connect to broker
    client.loop_start() #start the loop
    print("Subscribing to topic","humidity/alert")
    client.subscribe("humidity/alert")
    print("Publishing message to topic","humidity/alert")
    client.publish("humidity/alert",key["Humidity"])
    time.sleep(4) # wait
    client.loop_stop() #stop the loop
    if (key["Temperature"] > 28 and key["Pressure"] > 870 and key["Humidity"] > 12):
      client.on_message=on_message #attach function to callback
      print("connecting to broker")
      client.connect(broker_address) #connect to broker
      client.loop_start() #start the loop
      print("Subscribing to topic","/alert")
      client.subscribe("pressure/alert")
      print("Publishing message to topic","/alert")
      client.publish("/alert","Alert")
      time.sleep(4) # wait
      client.loop_stop() #stop the loop
```
